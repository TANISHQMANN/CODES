```python
import pandas as pd

df=pd.read_excel("/content/Classes June 2018 Cleaned and preprocessed data.xlsx")
df1=pd.read_excel("/content/Cleaned and Preprocessed data Classes april-May 2018.xlsx")
```

```python
#visualizations that reveal booking trends between columns max bookees and number booked

import matplotlib.pyplot as plt

plt.scatter(df['MaxBookees'], df['Number Booked'])
plt.xlabel('Max Bookees')
plt.ylabel('Number Booked')
plt.title('Scatter Plot of Max Bookees vs Number Booked')
plt.show()

df.plot.scatter(x='MaxBookees', y='Number Booked')
plt.xlabel('Max Bookees')
plt.ylabel('Number Booked')
plt.title('Scatter Plot of Max Bookees vs Number Booked')
plt.show()
```

```python
import seaborn as sns

# Create a line chart of the number of bookings per BookingEndDateTime (Month / Day / Year)
sns.lineplot(data=df, x="BookingEndDateTime (Month / Day / Year)", y="Number Booked")
plt.xlabel("BookingEndDateTime (Month / Day / Year)")
plt.ylabel("Number of Bookings")
plt.title("Number of Bookings")
plt.show()


```

```python
# Create a bar chart of the average number of bookings per class type
df.groupby('MaxBookees')['Number Booked'].mean().plot(kind='bar')
plt.xlabel('Class Type')
plt.ylabel('Average Number of Bookings')
plt.title('Average Number of Bookings per Class Type')
plt.show()

# Create a pie chart of the number of bookings per class type
df['MaxBookees'].value_counts().plot(kind='pie')
plt.xlabel('Class Type')
plt.ylabel('Number of Bookings')
plt.title('Number of Bookings per Class Type')
plt.show()

```

```python
# Perform descriptive statistics to summarize data features (e.g., mean, median, standard deviation).

df.describe()
```

```python
# @title
# Identify outliers using statistical tests
import scipy.stats as stats

# Test for outliers in MaxBookees
z_scores, p_values = stats.zscore(df['MaxBookees']), stats.norm.sf(abs(stats.zscore(df['MaxBookees'])))
outliers = df[p_values < 0.05]

# Print the outliers
print('Outliers in MaxBookees:')
print(outliers)

# Test for outliers in Number Booked
z_scores, p_values = stats.zscore(df['Number Booked']), stats.norm.sf(abs(stats.zscore(df['Number Booked'])))
outliers = df[p_values < 0.05]

# Print the outliers
print('Outliers in Number Booked:')
print(outliers)
```
